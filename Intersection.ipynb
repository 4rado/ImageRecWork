```python
import math
import numpy as np

def VectFrom2Points(point1, point2):
    return[point2[0] - point1[0], point2[1] - point1[1]]

def VectMult(vect1, vect2):
    vm = vect1[0]*vect2[1] - vect1[1]*vect2[0]
    if(vm != 0):
        vm = vm / abs(vm)
    return vm
def PointIsInPolygon(inPoint, polygon):
    length = len(polygon)
    #Берем первую сторону прямоугольника
    vectOfPolygonSide = VectFrom2Points(polygon[length - 1], polygon[0])
    #Проводим вектор из проверяемой точки в 0 - вую
    vectFromPoint = VectFrom2Points(inPoint, polygon[0])
    #Проверочная константа(нормированная)
    check = VectMult(vectFromPoint, vectOfPolygonSide)
    if(check == 0):
        return True
    for i in range(1, len(polygon)):
        vectOfPolygonSide = VectFrom2Points(polygon[i-1], polygon[i])  
        vectFromPoint = VectFrom2Points(inPoint, polygon[i])
        if( VectMult(vectFromPoint, vectOfPolygonSide) == 0 ):
        if( check != VectMult(vectFromPoint, vectOfPolygonSide) ):
            return False
    return True
        
# Пересечение прямоугольников


#проверяет лежит ли точка в отрезке
def PointIsInLine(point, line){
    k1 = 200000000
    k2 = 200000000
    if(point[0] - line[0][0] != 0 ):
        k1 = (point[1] - line[0][1])/(point[0] - line[0][0])
    if(point[0] - line[1][0] != 0 ):
        k2 = (point[1] - line[1][1])/(point[0] - line[1][0])
    
    if( k1 == k2 and ( point[0] >= line[0][0] and point[0] <= line[1][0] or point[0] <= line[0][0] and point[0] >= line[1][0] ) ):
        return True
    return False
}
# Возвращает для отрезков лежащих на одной прямой их пересечение
def CommonPoints(line1, line2):
    x11 = line1[0][0]
    x12 = line1[1][0]
    x21 = line2[0][0]
    x22 = line2[1][0]
    if( x11 > x12 ):
        x11, x12 = x12, x11
    if( x21 > x22 ):
        x21, x22 = x22, x21
    points = [x11,x12,x21,x22]
    sorted(points)
    medPoint = (points[1] + points[2])/2
    if( medPoint >= x11 and medPoint >= x21 and medPoint <= x12 and medPoint <= x22 ):
        return [points[1],points[2]]
    return [nan,nan]
    
def LineIntersection(line1, line2):
    #Коэффициент наклона прямой
    k1 = (line1[0][1] - line1[1][1])/(line1[0][0] - line1[1][0])
    k2 = (line2[0][1] - line2[1][1])/(line2[0][0] - line2[1][0])
    
    #Константа
    с1 = line1[0][1] - k1*line1[0][0]
    c2 = line2[0][1] - k1*line2[0][0]
    
    #1случай прямые параллельны
    if(k1 == k2):
        if(c1 == c2):
            return CommonPoints(line1, line2)
        else:
            return [nan,nan]
    
    # Находим точку пересечения двух отрезков и проверяем лежит ли она в каждом из них
    else:
        x = (с2 - с1)/(k1 - k2);
        y = (k1*c2 - k2*c1)/(k1 - k2);
        point = [x,y]
        if( PointIsInLine(point,line1) and PointIsInLine(point,line2)):
            return [point,point]
        else:
            return [nan, nan]
        

def PolygonsIntersection(polygon1, polygon2):
    allPoints = []
    linesFromPolygon1 = []
    linesFromPolygon2 = []
    
    linesFromPolygon1.append( polygon1[len(polygon1) - 1], polygon1[0] )
    for i in range( len(polygon1) - 1 ):
        linesFromPolygon1.append( polygon1[i], polygon1[i + 1] )
    
    linesFromPolygon2.append( polygon2[len(polygon1) - 1], polygon2[0] )
    for i in range( len(polygon2) - 1 ):
        linesFromPolygon2.append( polygon2[i], polygon2[i + 1] )
    
    
    for line1 in linesFromPolygon1:
        for line2 in linesFromPolygon2:
            inters = LineIntersection( line1, line2 )
            if( !math.isnan(inters[0]) ):
                if( inters[0] == inters[1] ):
                    allPoints.append( inters[0] )
                else:
                    allPoints.append( inters[0] )
                    allPoints.append( inters[1] )

        
    
```
