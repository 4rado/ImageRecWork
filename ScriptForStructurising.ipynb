```python
import os
import sys
import datetime
 
date = datetime.datetime.now()
time = date.time()
stime = "" 

for ch in str(time)[0:8]:
    if(ch != ':'):
        stime += ch

path = os.path.abspath( "C:\Diplom\Markup_for_Adding2D" )
pathImage = os.path.abspath( "C:\Diplom\Markup_for_Adding2DCopy\Image" )
filesList = os.listdir(pathImage)
for file in filesList:
    name =  os.path.splitext(file)[0]
    ext = os.path.splitext(file)[1]
    newname = name + stime + ext
    name = name + ext
    print(newname)
    os.rename( os.path.join(pathImage, name ), os.path.join(pathImage, newname ) )
num = 1
fileList = os.listdir(pathImage)
for file in fileList:
    name =  os.path.splitext(file)[0]
    ext = os.path.splitext(file)[1]
    newname = "{:04}".format(num) + ext
    name = name + ext
    num += 1
    print(newname)
    os.rename( os.path.join(pathImage, name ), os.path.join(pathImage, newname ) )
    
    
    
```
